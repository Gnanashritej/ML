```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,r2_score

from sklearn.metrics import mean_squared_error
from sklearn import linear_model
from sklearn.linear_model import LogisticRegression
```

```python
data=pd.read_csv("auto-mpg.csv")
data=data.dropna(axis=0,how='any')
predictors=data.iloc[:,1:7] #seggregating the predictor variables...........
```

```python
data.head(5)
```

```python
data=data.dropna(axis=0,how='all')
data.head(5)
```

```python
target = data.iloc[:,0] #seggregating the target/class variable......
```

```python
predictors
```

```python
predictors.shape
```

```python
predictors.columns
```

```python
predictors_train, predictors_test, target_train, target_test = train_test_split(predictors,target,test_size = 0.3,random_state = 123)
```

```python
lm = linear_model.LinearRegression()
```

```python
predictors_train
```

```python
target_train
```

```python
#First train model/ classifier with the input dataset (training data part)
model = lm.fit(predictors_train, target_train)
```

```python
t_p = model.predict(predictors_train)
mean_squared_error(target_train, t_p)
r2_score(target_train,t_p)
```

```python
#Make prediction using the trained model
prediction = model.predict(predictors_test)
mean_squared_error(target_test,prediction)
r2_score(target_test,prediction)
```
